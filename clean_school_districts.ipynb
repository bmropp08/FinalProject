```python
import pandas as pd

# Replace 'file_path.csv' with the actual path to your CSV file.
file_path = 'Public_School_Characteristics_2020-21.csv'

# Use pandas.read_csv() to read the CSV file into a DataFrame.
# By default, the first row of the CSV will be treated as the header row.
# If your CSV file has no header, you can use `header=None` as an argument.
# If you want to specify a custom separator, you can use the `sep` parameter (e.g., sep=';').
data_frame = pd.read_csv(file_path)

# Now you can work with the data stored in the DataFrame.
# For example, you can display the first few rows using the `head()` method:
data_frame.head()
```

```python
#dropping multuple columns
columns_to_drop = ['OBJECTID', 'NCESSCH', 'SURVYEAR', 'LEAID', 'ST_LEAID', 'HIALF', 'HI', 'TRALM', 'TRALF', 'TR', 'WHALF', 'WHALM','WH']
data_frame = data_frame.drop(columns_to_drop, axis=1)
data_frame.head()
```

```python
columns_to_drop = ['LSTREET1', 'LSTREET2', 'AS', 'BLALM', 'BLALF', 'BL', 'HPALM', 'HPALF', 'HP', 'HIALM']
data_frame = data_frame.drop(columns_to_drop, axis=1)
data_frame.head()
```

```python
columns_to_drop = ['LZIP4', 'PHONE', 'MEMBER', 'FTE', 'STUTERATIO', 'AMALF', 'AM', 'ASALM', 'ASALF', 'LATCOD', 'LONCOD']
data_frame = data_frame.drop(columns_to_drop, axis=1)
data_frame.head()
```

```python
columns_to_drop = ['CHARTER_TEXT', 'MAGNET_TEXT', 'G10','G11', 'G12', 'G13', 'UG', 'AE', 'TOTMENROL', 'TOTFENROL', 'TOTAL', 'AMALM']
data_frame = data_frame.drop(columns_to_drop, axis=1)
data_frame.head()
```
